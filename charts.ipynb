```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/flowtime-den312d.csv')
plt.figure(figsize=(10, 6))
#plt.style.use('darkgrid')


plt.plot(df['agents'], df['TSWAP'], marker='o', linestyle='-', linewidth=2, label='TSWAP')
plt.plot(df['agents'], df['FlowNetwork'], marker='s', linestyle='--', linewidth=2, label='FlowNetwork')
plt.plot(df['agents'], df['GAP-1'], marker='^', linestyle='-.', linewidth=2, label='GAP-1')
plt.plot(df['agents'], df['GAP-50'], marker='*', linestyle=':', linewidth=2, label='GAP-50')
plt.plot(df['agents'], df['GAP-GRID'], marker='P', linestyle='-.', linewidth=2, label='GAP-GRID')

plt.xlabel('Number of Agents', fontsize=12)
plt.ylabel('Flowtime', fontsize=12)
plt.title('Flowtime for den312d', fontsize=14)


plt.legend(fontsize=10)
plt.grid(True, alpha=0.5)


plt.xticks(df['agents'], fontsize=8)
plt.xlim(10, 100)

plt.tight_layout()
plt.show()
```
